# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 描画の基本

```python
x = np.linspace(-2, 2, 100)
plt.plot(x, x**2)
```

## 複数グラフの描画

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, x**2, label='x**2')
axes[1].plot(x, x**3, 'orange', label='x**3')
axes[1].plot(x, x, label='x')
axes[1].legend()
axes[0].legend()
fig.tight_layout()
```

## 散布図：plt.scatter()

```python
df = pd.read_csv('../dataset/tmdb_5000_movies.csv')
```

```python
plt.scatter(df['revenue'], df['budget'], alpha=0.2)
plt.xlabel('revenue')
plt.ylabel('bugget')
plt.show()
```

```python
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
lang_group_df = df.groupby('original_language').mean()[['budget', 'revenue', 'vote_count']]
c = np.arange(len(lang_group_df))
plt.scatter('revenue', 'budget', s='vote_count', c=c, data=lang_group_df)
for idx, row in lang_group_df.iterrows():
    plt.annotate(idx, (row['revenue'], row['budget']))  ## plt.annotate('text', (x, y))
```

## ヒストグラム：plt.hist()

```python
plt.hist(df['vote_average'], bins=50)
plt.show()
```

```python
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
df['profit'] = df['revenue'] - df['budget']
df.head()
```

```python
fig, axes = plt.subplots()
logbins= np.logspace(0, np.log10(df['profit'].max()))
axes.hist(df['profit'], bins=logbins)
axes.set_xscale('log')
```
